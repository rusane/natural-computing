# Result Plots

Plots using 30 runs csv files of base learners and stacking.

```python
# import libraries
import warnings

warnings.filterwarnings(action='ignore')

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
sns.set(style='whitegrid', palette="deep", font_scale=0.9)
```

## Load Data

```python
# load base learners (with best params)
lra_df = pd.read_csv("./results/lra.csv")
dtc_df = pd.read_csv("./results/lra.csv")
ann_df = pd.read_csv("./results/ann.csv")
svm_df = pd.read_csv("./results/svc.csv")

# load base learners (with baseline params)
def_lra_df = pd.read_csv("./results/def_lra.csv")
def_dtc_df = pd.read_csv("./results/def_lra.csv")
def_ann_df = pd.read_csv("./results/def_ann.csv")
def_svm_df = pd.read_csv("./results/def_svc.csv")

# load base learners (with best params + refit data)
refit_lra_df = pd.read_csv("./results/refit_lra.csv")
refit_dtc_df = pd.read_csv("./results/refit_lra.csv")
refit_ann_df = pd.read_csv("./results/refit_ann.csv")
refit_svm_df = pd.read_csv("./results/refit_svc.csv")

# load base learners (with baseline params + refit data)
def_refit_lra_df = pd.read_csv("./results/def_refit_dtc.csv")
def_refit_dtc_df = pd.read_csv("./results/def_refit_lra.csv")
def_refit_ann_df = pd.read_csv("./results/def_refit_ann.csv")
def_refit_svm_df = pd.read_csv("./results/def_refit_svc.csv")

# load stacking
stk_df = pd.read_csv('./results/stk.csv')
stk_cv_df = pd.read_csv('./results/stk_cv.csv')
```

```python
def plot_result(df1, df2, df3, df4, y1, t1, y2, t2, y3, t3, y4, t4):
    '''function to visualize box plots'''
    plt.figure(figsize=(25,25))
    plt.subplot(2,2,1)
    sns.boxplot(x="metrics", y="value", 
                data=pd.melt(df1, var_name='metrics'))
    sns.swarmplot(x="metrics", y="value", color=".25",
                  data=pd.melt(df1, var_name='metrics'))
    plt.ylim(y1)
    plt.title(t1)
    
    plt.subplot(2,2,2)
    sns.boxplot(x="metrics", y="value", 
                data=pd.melt(df2, var_name='metrics'))
    sns.swarmplot(x="metrics", y="value", color=".25",
                  data=pd.melt(df2, var_name='metrics'))
    plt.ylim(y2)
    plt.title(t2)
    
    plt.subplot(2,2,3)
    sns.boxplot(x="metrics", y="value", 
                data=pd.melt(df3, var_name='metrics'))
    sns.swarmplot(x="metrics", y="value", color=".25",
                  data=pd.melt(df3, var_name='metrics'))
    plt.ylim(y3)
    plt.title(t3)
    
    plt.subplot(2,2,4)
    sns.boxplot(x="metrics", y="value", 
                data=pd.melt(df4, var_name='metrics'))
    sns.swarmplot(x="metrics", y="value", color=".25",
                  data=pd.melt(df4, var_name='metrics'))
    plt.ylim(y4)
    plt.title(t4)
    
    plt.show()
```

### Base Learners

#### LRA

```python
plot_result(def_lra_df, lra_df, def_refit_lra_df, refit_lra_df,
            [0.7,1.1], "LRA (Baseline)",
            [0.7,1.1], "LRA (Grid Search)",
            [0.7,1.1], "LRA (Baseline + Refit)",
            [0.7,1.1], "LRA (Grid Search + Refit)")
```

#### DT

```python
plot_result(def_dtc_df, dtc_df, def_refit_dtc_df, refit_dtc_df,
            [0.7,1], "DT (Baseline)",
            [0.7,1], "DT (Grid Search)",
            [0.7,1], "DT (Baseline + Refit)",
            [0.7,1], "DT (Grid Search + Refit)")
```

#### ANN

```python
plot_result(def_ann_df, ann_df, def_refit_ann_df, refit_ann_df,
            [0.7,1], "ANN (Baseline)",
            [0.7,1], "ANN (Grid Search)",
            [0.7,1], "ANN (Baseline + Refit)",
            [0.7,1], "ANN (Grid Search + Refit)")
```

#### SVM

```python
plot_result(def_svm_df, svm_df, def_refit_svm_df, refit_svm_df,
            [0.7,1], "SVM (Baseline)",
            [0.7,1], "SVM (Grid Search)",
            [0.7,1], "SVM (Baseline + Refit)",
            [0.7,1], "SVM (Grid Search + Refit)")
```

### Stacking

```python
plot_result(stk_df, stk_cv_df, 
            [0.8,1], "Stacking (Best Base Learner Params)",
            [0.8,1], "Stacking (Grid Search Params)")
```
